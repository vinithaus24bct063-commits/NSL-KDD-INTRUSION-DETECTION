```
import zipfile
import os

# Assuming zip_file_name contains the name of your uploaded zip file
# For example: zip_file_name = 'DrDoS_DNS.csv (2).zip'

# Extract the CSV from the zip file
with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall('.')

# Assuming the CSV file inside the zip has the same base name as the zip file
# without the .zip extension and potential '(2)'
csv_file_name_from_zip = zip_file_name.replace(' (2).zip', '.csv').replace('.zip', '.csv')

# Check if the extracted file exists, if not, try to list contents of zip
if not os.path.exists(csv_file_name_from_zip):
    print(f"Could not find {csv_file_name_from_zip} after extraction. Listing zip contents...")
    with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
        zip_contents = zip_ref.namelist()
        print("Zip contents:", zip_contents)
        if csv_contents := [f for f in zip_contents if f.endswith('.csv')]:
            csv_file_name_from_zip = csv_contents[0]
            print(f"Using first CSV found in zip: {csv_file_name_from_zip}")
        else:
            raise FileNotFoundError("No CSV file found in the zip archive.")

df = pd.read_csv(csv_file_name_from_zip)
print(df.head())
```
